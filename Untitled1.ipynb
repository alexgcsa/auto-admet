```python
import pandas as pd
import numpy as np

# Random data generation for demonstration (e.g., 100 samples, 3 features)
np.random.seed(0)
data = pd.DataFrame(np.random.randint(0, 2, size=(100, 4)), columns=['X1', 'X2', 'X3', 'y'])
```

```python
from pgmpy.estimators import HillClimbSearch, BicScore

# Initialize the Hill Climb Search
hc = HillClimbSearch(data, scoring_method=BicScore(data))

# The best model estimation
best_model = hc.estimate()
print("Edges:", best_model.edges())
```
